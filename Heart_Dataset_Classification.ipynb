## Data cleaning

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['RestBP'] = df['RestBP'].fillna(df['RestBP'].mean())
```

```python
df['Ca'] = df['Ca'].fillna(1).astype(int)
df['Ca'].values
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['AHD']=df['AHD'].replace({'Yes':1, 'No':0})
df.head()
```

## Data Transformation

```python
df['ChestPain'].nunique()
```

```python
df['Thal'].nunique()
```

### Splitting dataset into training and testing

```python
from sklearn.model_selection import train_test_split
y = df.pop('AHD')
y = np.array(y)
```

```python
X = df
X
```

```python
type(X)
```

```python
type(y)
```

### Encoding independent features using OneHotEncoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [2,12])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
print(X[0])
```

```python
type(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(np.concatenate((y_pred.reshape(-1,1),y_test.reshape(-1,1)),1))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cf_matrix = confusion_matrix(y_test,y_pred)
print(cf_matrix)
```

```python
print('Accuracy of the model:',accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
