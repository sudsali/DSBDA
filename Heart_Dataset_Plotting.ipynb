```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/Heart.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.dtypes
```

## Data cleaning

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['RestBP'] = df['RestBP'].fillna(df['RestBP'].mean())
```

```python
df['Ca'] = df['Ca'].fillna(1).astype(int)
df['Ca'].values
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['AHD']=df['AHD'].replace({'Yes':1, 'No':0})
df.head()
```

## Graphs

### People having heart disease

```python
sns.countplot(x='AHD', data=df)
plt.show()
```

### Number of males and females

```python
sns.countplot(x='Sex', data=df)
plt.ylabel('Number of people')
plt.show()
```

### Relation of age and Resting BP

```python
sns.relplot(x='Age', y='RestBP', data=df)
plt.title('Age vs Resting BP')
plt.show()
```

### Chest pain types

```python
sns.countplot(x='ChestPain', data=df)
plt.title("Chest Pain Types")
plt.xlabel('Chest Pain Type')
plt.ylabel('Number of patients')
plt.show()
```

### Resting BP vs Gender

```python
sns.catplot(x='Sex', y='RestBP', data=df)
plt.title('Gender vs Resting BP')
plt.show()
```

### Resting EG types

```python
df['RestECG'].value_counts().plot.bar()
plt.title('Resting ECG types')
plt.xlabel('Types of Resting ECG')
plt.ylabel('Number of people')
plt.show()
```

### Slope types

```python
plt.pie(df['Slope'].value_counts(), labels=['3','2','1'], autopct='%1.1f%%', shadow=True)
plt.title('Ratios of Slope Types')
plt.show()
```

### Cholestrol

```python
plt.hist(df['Chol'])
plt.xlabel('Cholestrol')
plt.show()
```
