# Data preprocessing

```python
fb_data = pd.read_csv('/content/sample_data/dataset_Facebook.csv',';')
fb_data
```

Checking for null values in the dataset

```python
fb_data.isnull().sum()
```

```python
# delete rows with null values
fb_data.dropna(axis=0,inplace=True)
fb_data
```

```python
# check for null values again (confirm that there are no null values)
fb_data.isnull().sum()
```

Thus, all null values are removed

# Creating data subsets -

```python
print(fb_data.columns)
```

```python
fb_data.shape
```

### Creating subsets row-wise

```python
row_subset_1 = fb_data.loc[:245,:]
row_subset_1
```

```python
row_subset_2 = fb_data.loc[250:,:]
row_subset_2
```

### Creating subsets column-wise

```python
col_subset_1 = fb_data.iloc[:,:10]
col_subset_1
```

```python
col_subset_2 = fb_data.iloc[:,11:]
col_subset_2
```

# Merging data subsets created above

### Row merge

```python
row_merge = pd.concat([row_subset_1, row_subset_2])
row_merge
```

### Column merge

```python
col_merge = pd.merge(col_subset_1, col_subset_2, how='left', left_index=True, right_index=True)
col_merge
```

# Sorting data

```python
sort_by_like = fb_data.sort_values(by=['like']) # By default, ascending is True
sort_by_like
```

```python
sort_by_like_desc = fb_data.sort_values(by=['like'], ascending=False)
sort_by_like_desc
```
